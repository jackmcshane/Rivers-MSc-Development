# Augmentation Code

```python
# Referenced from https://albumentations.ai/docs/getting_started/image_augmentation/

# Setup the Augmentation Pipeline, which will decide the parameters that will be applied to the waste dataset
# Pipeline
wastetransform = A.Compose([
    A.HorizontalFlip(p=1),
    A.Rotate(limit=360, border_mode=0, p=1),
    A.RandomBrightnessContrast(brightness_limit=(-0.2, 0.1), contrast_limit=(-0.2, 0.1), always_apply=False, p=0.5),
    A.MedianBlur(blur_limit=(3), p=0.3),
    A.Flip(p=0.5),
    A.HueSaturationValue (hue_shift_limit=20, sat_shift_limit=20, val_shift_limit=20, always_apply=False, p=0.3),
    A.JpegCompression (quality_lower=40, quality_upper=100, always_apply=False, p=0.5),
    A.MotionBlur(blur_limit=(3), p=0.3),
    A.RGBShift (r_shift_limit=20, g_shift_limit=20, b_shift_limit=20, always_apply=False, p=0.5),
    A.RandomGamma(gamma_limit=(50, 200), always_apply=False, p=0.5),
])
# Go through each folder and augment the 100x100 dataset.
# Iteration 1
for x in range(0,9):
  
    # Read relevant waste image folder
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_00" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)
      
    # Augment each image X times  
    for i in range (12):
      #Perform Augmentation
      clearBottle_aug = wastetransform(image = waste_rgb_bottleclear)

      # Write new adapted images
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/GPU ALB Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio + Augmentation)(V2))/Clear Plastic Bottle/" + "CP_00" + str(x+1) + "Augment" + str(i+1) + ".jpg", 
                  cv2.cvtColor(clearBottle_aug["image"], cv2.COLOR_BGR2RGB))
      
      # Print Images
      print(str(x+1) + " Augmented Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(clearBottle_aug["image"])
      plt.show()

    else:
      print("Empty Frame")
      continue

# Iteration 2
for x in range(9,99):
  
    # Read relevant waste image folder
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_0" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)

    # Augment each image X times
    for i in range (12):
      # Perform Augmentation
      clearBottle1_aug = wastetransform(image = waste_rgb_bottleclear)
      
      # Write new adapted images
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/GPU ALB Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio + Augmentation)(V2))/Clear Plastic Bottle/" + "CP_0" + str(x+1) +"Augment" + str(i+1) + ".jpg", cv2.cvtColor(clearBottle1_aug["image"], cv2.COLOR_BGR2RGB))
      
      # Print Images
      print(str(x+1) + " Augmented Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(clearBottle1_aug["image"])
      plt.show()
      
    else:
      print("Empty Frame")
      continue

# Iteration 3
for x in range(99,300):
  
    # Read relevant waste image folder
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)
      
    # Augment each image X times  
    for i in range (12):
      # Perform Augmentation
      clearBottle2_aug = wastetransform(image = waste_rgb_bottleclear)

      # Write new adapted Images
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/GPU ALB Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio + Augmentation)(V2))/Clear Plastic Bottle/" + "CP_" + str(x+1) + "Augment" + str(i+1) + ".jpg", cv2.cvtColor(clearBottle2_aug["image"], cv2.COLOR_BGR2RGB))
      
      # Print images
      print(str(x+1) + " Augmented Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(clearBottle2_aug["image"])
      plt.show()
      
    else:
      print("Empty Frame")
      exit(1)
```
