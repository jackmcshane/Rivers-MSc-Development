# Changing Aspect Ratio and Size of images code


```python
# Gain Back some Aspect ratio
inter = cv2.INTER_AREA

# New Sizes for images
width = 100
height = 100

# Resize all Clear Plastic Bottles images to 100 x 100 and add interpolation aspect ratio adjustment
# Note: Each waste class had to be broken up into three iteration loops to avoid displaying empty frame due to the class images being led by 0's
# EG. CPB_001, CPB_010, CPB_100

# Iteration of Clear Plastic Bottle 1
for x in range(0,9):
  
    # Read Clear Plastic Bottle Class folder in original format
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bottle/" + "CP_00" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)
      
      # Set the new dimensions to  variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio
      waste_rgb_newsize_1_12 = cv2.resize(waste_rgb_bottleclear, new_dimensions, interpolation = inter)

      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_00" + str(x+1) + ".jpg", 
                  cv2.cvtColor(waste_rgb_newsize_1_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_1_12)
      plt.show()

    # Else images not available
    else:
      print("Empty Frame")
      continue

# Iteration of Clear Plastic Bottle 2
for x in range(9,99):
  
    # Read Clear Plastic Bottle Class folder in original format
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bottle/" + "CP_0" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)
      
      # Set the new dimensions to  variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio
      waste_rgb_newsize_2_12 = cv2.resize(waste_rgb_bottleclear, new_dimensions, interpolation = inter)

      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_0" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_2_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_2_12)
      plt.show()
    # Else image not available 
    else:
      print("Empty Frame")
      continue

# Iteration of Clear Plastic Bottle 3
for x in range(99,300):
  
    # Read Clear Plastic Bottle Class folder in original format
    waste_JPG_bottleclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bottle/" + "CP_" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottleclear to RGB colour space
    if waste_JPG_bottleclear is not None:
      waste_rgb_bottleclear = cv2.cvtColor(waste_JPG_bottleclear, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio
      waste_rgb_newsize_3_12 = cv2.resize(waste_rgb_bottleclear, new_dimensions, interpolation = inter)

      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bottle/" + "CP_" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_3_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_3_12)
      plt.show()
      
    # Else image not available  
    else:
      print("Empty Frame")
      continue

# Iteration of Coloured Plastic Bottle 1
for x in range(0,9):
  
    # Read Coloured Plastic Bottle Class folder in original format
    waste_JPG_bottlecolour = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bottle/" + "CPB_00" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottlecolour to RGB colour space
    if waste_JPG_bottlecolour is not None:
      waste_rgb_bottlecolour = cv2.cvtColor(waste_JPG_bottlecolour, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_4_12 = cv2.resize(waste_rgb_bottlecolour, new_dimensions, interpolation = inter)

      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bottle/" + "CPB_00" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_4_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_4_12)
      plt.show()

    # Else Image not available
    else:
      print("Empty Frame")
      continue

# Iteration of Coloured Plastic Bottle 2
for x in range(9,99):
  
    # Read Coloured Plastic Bottle Class folder in original format
    waste_JPG_bottlecolour = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bottle/" + "CPB_0" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottlecolour to RGB colour space
    if waste_JPG_bottlecolour is not None:
      waste_rgb_bottlecolour = cv2.cvtColor(waste_JPG_bottlecolour, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_5_12 = cv2.resize(waste_rgb_bottlecolour, new_dimensions, interpolation = inter)
      
      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bottle/" + "CPB_0" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_5_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_5_12)
      plt.show()

    # Else image not available  
    else:
      print("Empty Frame")
      continue

# Iteration of Coloured Plastic Bottle 3

for x in range(99,300):
  
    # Read Coloured Plastic Bottle Class folder in original format
    waste_JPG_bottlecolour = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bottle/" + "CPB_" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_bottlecolour to RGB colour space
    if waste_JPG_bottlecolour is not None:
      waste_rgb_bottlecolour = cv2.cvtColor(waste_JPG_bottlecolour, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_6_12 = cv2.resize(waste_rgb_bottlecolour, new_dimensions, interpolation = inter)
      
      # Write adapted images to new directory      
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bottle/" + "CPB_" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_6_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_6_12)
      plt.show()

    # Else image not available  
    else:
      print("Empty Frame")
      continue


# Iteration of Clear Plastic Bag 1
for x in range(0,9):
  
    # Read Clear Plastic Bags Class folder in original format
    waste_JPG_Bagclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bags/" + "CPBB_00" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_Bagclear to RGB colour space
    if waste_JPG_Bagclear is not None:
      waste_rgb_Bagclear = cv2.cvtColor(waste_JPG_Bagclear, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_7_12 = cv2.resize(waste_rgb_Bagclear, new_dimensions, interpolation = inter)
      
      # Write adapted images to new directory    
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bags/" + "CPBB_00" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_7_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_7_12)
      plt.show()
      
    # Else Image not available  
    else:
      print("Empty Frame")
      continue

# Iteration of Clear Plastic Bag 2
for x in range(9,99):
  
    # Read Clear Plastic Bags Class folder in original format
    waste_JPG_Bagclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bags/" + "CPBB_0" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_Bagclear to RGB colour space
    if waste_JPG_Bagclear is not None:
      waste_rgb_Bagclear = cv2.cvtColor(waste_JPG_Bagclear, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_8_12 = cv2.resize(waste_rgb_Bagclear, new_dimensions, interpolation = inter)

      # Write adapted images to new directory    
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bags/" + "CPBB_0" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_8_12, cv2.COLOR_BGR2RGB))
      
      # Print detail of each image and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_8_12)
      plt.show()
      
    # Else image not available  
    else:
      print("Empty Frame")
      continue

# Iteration of Clear Plastic Bag 3
for x in range(99,200):
  
    # Read Clear Plastic Bag Class folder in original format
    waste_JPG_Bagclear = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Clear Plastic Bags/" + "CPBB_" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_Bagclear to RGB colour space
    if waste_JPG_Bagclear is not None:
      waste_rgb_Bagclear = cv2.cvtColor(waste_JPG_Bagclear, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_9_12 = cv2.resize(waste_rgb_Bagclear, new_dimensions, interpolation = inter)
      
      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Clear Plastic Bags/" + "CPBB_" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_9_12, cv2.COLOR_BGR2RGB))
      
      # Print image details and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_9_12)
      plt.show()
    
    # Else image not available
    else:
      print("Empty Frame")
      continue

# Iteration of Coloured Plastic Bag 1

for x in range(0,9):
  
    # Read Coloured Plastic Bag Class folder in original format
    waste_JPG_Colouredbag = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bags/" + "MPB_00" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_ColouredBag to RGB colour space
    if waste_JPG_Colouredbag is not None:
      waste_rgb_Colouredbag = cv2.cvtColor(waste_JPG_Colouredbag, cv2.COLOR_BGR2RGB)
      
      # Set new dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_10_12 = cv2.resize(waste_rgb_Colouredbag, new_dimensions, interpolation = inter)
      
      # Write adapted images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bags/" + "MPB_00" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_10_12, cv2.COLOR_BGR2RGB))
      
      # Print images details and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_10_12)
      plt.show()

    # Else Image not available
    else:
      print("Empty Frame")
      continue 

# Iteration of Coloured Plastic Bag 2

for x in range(9,99):
  
    # Read Coloured Plastic Bag Class folder in original format
    waste_JPG_Colouredbag = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bags/" + "MPB_0" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_ColouredBag to RGB colour space
    if waste_JPG_Colouredbag is not None:
      waste_rgb_Colouredbag = cv2.cvtColor(waste_JPG_Colouredbag, cv2.COLOR_BGR2RGB)
      
      # Set dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_11_12 = cv2.resize(waste_rgb_Colouredbag, new_dimensions, interpolation = inter)

      # Write new images to a directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bags/" + "MPB_0" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_11_12, cv2.COLOR_BGR2RGB))
      
      # Print image details and image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_11_12)
      plt.show()

    # Else image not available
    else:
      print("Empty Frame")
      continue

# Iteration of Coloured Plastic Bag 3

for x in range(99,200):
  
    # Read Coloured Plastic Bag Class folder in original format
    waste_JPG_Colouredbag = cv2.imread("/content/gdrive/My Drive/Rivers V2/Datasets/Original Unmodified Dataset/Coloured Plastic Bags/" + "MPB_" + str(x+1) + ".jpg")
  
    # Convert waste_JPG_ColouredBag to RGB colour space
    if waste_JPG_Colouredbag is not None:
      waste_rgb_Colouredbag = cv2.cvtColor(waste_JPG_Colouredbag, cv2.COLOR_BGR2RGB)
      
      # Set dimensions to a variable
      new_dimensions = (width, height)

      # Take the iteration loop of images after conversion to RGB and assign it to its new dimensions and aspect ratio      
      waste_rgb_newsize_12_12 = cv2.resize(waste_rgb_Colouredbag, new_dimensions, interpolation = inter)
      
      # Write images to new directory
      cv2.imwrite("/content/gdrive/My Drive/Rivers V2/Datasets/Waste Dataset - Pre-processing (image resizing 100x100 (Aspect Ratio) (V2))/Coloured Plastic Bags/" + "MPB_" + str(x+1) + ".jpg", cv2.cvtColor(waste_rgb_newsize_12_12, cv2.COLOR_BGR2RGB))
      
      # Print image details and the image itself
      print(str(x+1) + " Waste Particle " + str(x+1) + " Size == 100x100 ")
      plt.imshow(waste_rgb_newsize_12_12)
      plt.show()
      
    # Else image not available 
    else:
      print("Empty Frame")
      continue
      exit(1)





    
```
